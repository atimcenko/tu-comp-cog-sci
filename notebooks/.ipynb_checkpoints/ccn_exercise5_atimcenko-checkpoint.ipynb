### Exercise 5.1

What is the BIC of a random policy (picking both options with equal probability) for this data-set?

Answer:

The Bayesian Information Criterion given the model is:
$$ BIC = |\Theta| logKT - 2  \underset{\Theta}{max}\ log\ p\left(\mathcal{D}|M=m, \Theta \right)$$where

$$ log\ p\left(\mathcal{D}|M=m, \Theta \right) = \sum_{k, t} log\ p\left(c_{t, k} | x_{t, k}, \Theta, m \right)$$

For random policy $|\Theta| = 0$ and the probability of choice is $p = 1/2$, hence for $m$ - model of random policy:

$$ BIC(m) = 0 - 2 \cdot \sum_{k, t} log 0.5 = -2 \cdot 15 \cdot 10 \cdot log 0.5 \approx 207.9$$

#### Define Learner classes

```python
# define models (random policy does not estimate anything, it will appear in choices)
class RescorlaWagner:
    name = "Rescorla Wagner"
    n_params = 2
    def __init__(self, num_inputs, learning_rate, epsilon):
        self.name = RescorlaWagner.name
        self.num_inputs = num_inputs
        self.learning_rate = learning_rate
        self.weights = np.zeros((num_inputs, 1))
        self.epsilon = epsilon
        
    def predict(self, inputs):
        return self.weights.T @ inputs
    
    def choose(self, pred_A, pred_B):
        """The inputs are the outputs of "predict" method for objects A and B"""
        p_A = int(pred_A > pred_B) * (1 - self.epsilon) + 0.5 * self.epsilon
        return stats.bernoulli(p=p_A).rvs()
             
    def learn(self, inputs, targets):
        self.weights = self.weights + self.learning_rate * (targets - self.predict(inputs)) * inputs
        
        
    def negloglike(self, pred_A, pred_B, choice):
        p_A = int(pred_A > pred_B) * (1 - self.epsilon) + 0.5 * self.epsilon
        return -np.log(stats.bernoulli(p=p_A).pmf(choice))
           

class KalmanFilter:
    name = "Kalman Filter"
    n_params = 3
    def __init__(self, num_inputs, sigma_y, sigma_w, epsilon):
        self.name = KalmanFilter.name
        self.num_inputs = num_inputs
        self.sigma_y = sigma_y
        self.sigma_w = sigma_w
        self.epsilon = epsilon
        
        self.weights = np.zeros((num_inputs, 1))
        self.covariance = sigma_w * np.eye(num_inputs)
        
    def predict(self, inputs):
        mean = self.weights.T @ inputs
        std = np.sqrt(inputs.T @ self.covariance @ inputs + self.sigma_y ** 2)
        return mean
    
    def choose(self, pred_A, pred_B):
        """The inputs are the outputs of "predict" method for objects A and B"""
        p_A = int(pred_A > pred_B) * (1 - self.epsilon) + 0.5 * self.epsilon
        return stats.bernoulli(p=p_A).rvs()
               
    def learn(self, inputs, targets):
        kalman_numerator = self.covariance @ inputs
        kalman_denominator = inputs.T @ self.covariance @ inputs + self.sigma_y ** 2
        kalman_gain = kalman_numerator / kalman_denominator
        
        self.weights = self.weights + kalman_gain * (targets - self.weights.T @ inputs)
        self.covariance = self.covariance - kalman_gain @ inputs.T @ self.covariance
        
    def negloglike(self, pred_A, pred_B, choice):
        p_A = int(pred_A > pred_B) * (1 - self.epsilon) + 0.5 * self.epsilon
        return -np.log(stats.bernoulli(p=p_A).pmf(choice))
        
        
class RandomModel: 
    name = "Random Model"
    n_params = 1
    def __init__(self, num_inputs, p):
        self.name = RandomModel.name
        self.p = p
        
    def predict(self, inputs):
        return inputs
    
    def choose(self, pred_A, pred_B):
        return stats.bernoulli(p=self.p).rvs()
    
    def learn(self, inputs, targets):
        pass
    
    def negloglike(self, pred_A, pred_B, choice):
        return -np.log(stats.bernoulli(p=self.p).pmf(choice))
```

#### The algorithm of finding BIC

The main component is maximum-likelihood estimate of the parameters. Rescorla Wagner model has two parameters - learning rate $\alpha$ and $\epsilon$. Kalman filter has three parameters - $\sigma_y$, $\sigma_w$ and $\epsilon$. For random model Bernoulli probability $p$ can be regarded as a parameter (for MLE not to be booring).

**MLE estimation loop**

For each model:
- Initialize grids (grid-search) or ranges (Monte-Carlo sampling) for parameters.

    For each participant:
    
    - Compute loglikelihood of choices given the parameters <- train model like usual using tasks and trials within the task.
    - Find the best parameter set which maximizes log-likelihood. Save the log-likelihood value.
    - Compute BIC using static task/model variables (# trials, # tasks, # parameters). Save BIC value.


#### Grid-search MLE

```python
%%time
np.random.seed(42)
VERBOSE = False

# initialize dict to store BIC values for each model and participant
model_names = ["Rescorla Wagner", "Kalman Filter", "Random Model"]
BIC = {k: np.empty(n_subjects) for k in model_names}
LL = {k: np.empty(n_subjects) for k in model_names}

# initialize param space for each model
n_gridpoints = 8

lr_grid = np.linspace(0.05, 0.8, n_gridpoints)
eps_grid = np.linspace(0.01, 0.5, n_gridpoints)
sigma_y_grid = np.linspace(0.1, 3, n_gridpoints)
sigma_w_grid = np.linspace(0.1, 3, n_gridpoints)
p_grid = np.linspace(0.05, 0.95, n_gridpoints)

model_params = {"Rescorla Wagner": list(product(lr_grid, eps_grid)), 
                "Kalman Filter": list(product(sigma_y_grid, sigma_w_grid, eps_grid)), 
                "Random Model": [(p, ) for p in p_grid]}

print("~~ GRID SEARCH MLE ~~ \n")

# Master loop though the models
for ModelClass in [KalmanFilter, RescorlaWagner, RandomModel]:
    
    if VERBOSE: 
        print("\n---------------------------")
        print("---------------------------")
        print(f"Fitting parameters for {ModelClass.name}")
        
    params = model_params[ModelClass.name]
    
    # Subject loop: fit different MLE params to different subjects
    for subject in range(n_subjects):
        if VERBOSE:
            print("\n---------------------")
            print(f"Subject #{subject} \n")

        inputs = data['inputs'][subject]
        targets = data['targets'][subject]
        choices = data['choices'][subject]
        
        min_nll = np.inf
        opt_param = ( )

        # MLE estimation loop
        for param in params:
            
            pred_choices = np.empty_like(choices)
            pred_targets = np.empty_like(targets)
            nll = np.empty_like(choices)
            
            for task in range(n_tasks):
                model = ModelClass(n_features, *param)
                for trial in range(n_trials):
                    pred_A, pred_B = model.predict(inputs[task, trial].T)[0]
                    pred_targets[task, trial, :] = pred_A, pred_B
                    
                    pred_choices[task, trial] = model.choose(pred_A, pred_B)
                    nll[task, trial] = model.negloglike(pred_A, pred_B, choices[task, trial]) 
                    
                    model.learn(inputs[task, [trial], 0, :].T, targets[task, [trial], 0])
                    model.learn(inputs[task, [trial], 1, :].T, targets[task, [trial], 1])
            
            total_nll = np.sum(nll)
            
            if total_nll < min_nll:
                min_nll = total_nll
                opt_param = param
                
        if VERBOSE: print("Optimal MLE parameters: ", np.round(np.array(opt_param), 2))
        
        # get BIC, record BIC and LL
        bic = ModelClass.n_params * np.log(n_tasks * n_trials) + 2 * min_nll    # "+" since calculated negative log-likelihood
        BIC[ModelClass.name][subject] = bic
        LL[ModelClass.name][subject] = -min_nll
        
        if VERBOSE: print(f"BIC = {bic:.1f}")
        if VERBOSE: print(f"Log-likelihood = {-min_nll:.1f}")
                    
np.save("../data/bic_grid.npy", BIC)
np.save("../data/ll_grid.npy", LL)
print("Grid-search MLE estimation completed.")
```

#### Monte-Carlo search MLE

```python
%%time
# for Monte-Carlo sampling
VERBOSE = False

def sample_MonteCarlo(param_bounds:list):
    param = [stats.uniform(loc=b[0], scale=b[1] - b[0]).rvs() for b in param_bounds]
    return param

model_names = ["Rescorla Wagner", "Kalman Filter", "Random Model"]

model_param_bounds = {"Rescorla Wagner": [(lr_grid[0], lr_grid[-1]), (eps_grid[0], eps_grid[1])],
                      "Kalman Filter": [(sigma_y_grid[0], sigma_y_grid[-1]), (sigma_w_grid[0], sigma_w_grid[-1]), (eps_grid[0], eps_grid[1])], 
                      "Random Model": [(p_grid[0], p_grid[-1])]}

BIC = {k: np.empty(n_subjects) for k in model_names}
LL = {k: np.empty(n_subjects) for k in model_names}

n_mc_samples = 32

print("~~ MONTE-CARLO SAMPLING MLE ~~ \n")

# Master loop though the models
for ModelClass in [KalmanFilter, RescorlaWagner, RandomModel]:
    
    if VERBOSE:
        print("\n---------------------------")
        print("---------------------------")
        print(f"Fitting parameters for {ModelClass.name}")
        
    param_bounds = model_param_bounds[ModelClass.name]
    
    # Subject loop: fit different MLE params to different subjects
    for subject in range(n_subjects):
        
        if VERBOSE:
            print("\n---------------------")
            print(f"Subject #{subject} \n")

        inputs = data['inputs'][subject]
        targets = data['targets'][subject]
        choices = data['choices'][subject]
        
        min_nll = np.inf
        opt_param = []

        # MLE estimation loop - the number of estimation points scales linearily with dimension
        for i_param in range(n_mc_samples * ModelClass.n_params):
            param = sample_MonteCarlo(param_bounds)
            
            pred_choices = np.empty_like(choices)
            pred_targets = np.empty_like(targets)
            nll = np.empty_like(choices)
            
            for task in range(n_tasks):
                model = ModelClass(n_features, *param)
                for trial in range(n_trials):
                    pred_A, pred_B = model.predict(inputs[task, trial].T)[0]
                    pred_targets[task, trial, :] = pred_A, pred_B
                    
                    pred_choices[task, trial] = model.choose(pred_A, pred_B)
                    nll[task, trial] = model.negloglike(pred_A, pred_B, choices[task, trial]) 
                    
                    model.learn(inputs[task, [trial], 0, :].T, targets[task, [trial], 0])
                    model.learn(inputs[task, [trial], 1, :].T, targets[task, [trial], 1])
            
            total_nll = np.sum(nll)
            
            if total_nll < min_nll:
                min_nll = total_nll
                opt_param = param
                
        if VERBOSE: print("Optimal MLE parameters: ", np.round(np.array(opt_param), 2))
        
        # get BIC. Record BIC and LL.
        bic = ModelClass.n_params * np.log(n_tasks * n_trials) + 2 * min_nll    # "+" since calculated negative log-likelihood
        BIC[ModelClass.name][subject] = bic
        LL[ModelClass.name][subject] = -min_nll
        
        if VERBOSE: print(f"Log-likelihood = {-min_nll:.1f}")
        if VERBOSE: print(f"BIC = {bic:.1f}")

np.save("../data/bic_mc.npy", BIC)
np.save("../data/ll_mc.npy", LL)
print("Monte-Carlo sampling MLE estimation completed.")
```

```python
model_names = ["Rescorla Wagner", "Kalman Filter", "Random Model"]
bic_arr = np.vstack([bic[model] for model in model_names])

# for each row will have 1 if the model wins and 0 otherwise
winners = bic_arr - bic_arr.min(axis=0)
winners[winners != 0] = 1
winners = 1 - winners

fig, axes = plt.subplots(2, 1, figsize=(6, 8))

ax = axes[0]
df = pd.DataFrame({model:winstreak for model, winstreak in zip(model_names, winners)})
seaborn.heatmap(df.T, ax=ax, cmap='gray_r', square=True, cbar=False)
ax.set_title("Winner model by best BIC for each participant. \n Win is indicated by a black square")
ax.set_xlabel("Participant #")

ax = axes[1]
ax.set_title("Left: Frequency of choice as the best explanatory model \nRight: summed BIC values across participants")

color1 = "red"
ax.bar(model_names, np.mean(winners, axis=1) * 100, color=color1, alpha=0.8, hatch=r"o")
ax.set_ylabel("% participants")

# summed BIC
# txt = ''.join(["Summed BIC"] + [f"\n{model}: {bic[model].sum():.0f}" for model in model_names])
# props = dict(boxstyle='round', facecolor='gray', alpha=0.2)
# ax.text("Kalman Filter", 40, txt, fontdict={'fontsize': 12}, bbox=props)

# histogram of summed BIC values
color2 = "#4287f5"
ax2 = ax.twinx()
ax2.bar(model_names, bic_arr.sum(axis=1), color=color2, alpha=0.3, hatch=r"//")
ax2.set_ylabel("Summed BIC", color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

plt.tight_layout()
```

### Exercise 5.3

Use the summed BIC values from the last exercise to approximate posterior probabilities over models. You may assume a uniform prior over models. Report the resulting posterior probabilities.

```python
bic_summed = bic_arr.sum(axis=1)

# logsumexp trick
c = np.max(-bic_summed / 2)
logsumexp = c + np.log(np.sum(np.exp(-bic_summed / 2 - c)))
# since priors is uniform don't need to account for that in the Bayes update formula (they cancel out)
logposteriors = -bic_summed / 2 - logsumexp
posteriors = np.exp(logposteriors)

print("Posterior probabilites:\n")
for model, prob in zip(model_names, posteriors):
    print(f"{model}: {prob:.1e}")
```
